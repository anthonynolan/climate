```python
import statsmodels.api as sm
import statsmodels.formula.api as smf
import matplotlib.pyplot as plt
```

```python
data = sm.datasets.get_rdataset('Duncan', 'carData', cache=True)
```

```python
results = smf.ols(formula='prestige ~ income', data=data.data).fit()
results.summary()
```
