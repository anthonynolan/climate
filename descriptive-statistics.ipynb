```python
from sklearn.datasets import load_diabetes
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = load_diabetes(as_frame=True, scaled=False)
print(data.DESCR)
```

```python
df = data.frame
df.describe()
```

```python
sns.displot(df.age, height=4, aspect=3, bins=20)
median = df.describe().loc['50%', 'age']
plt.axvline(median, c='r', label=f'median {median}')
plt.legend()
plt.show()
```
